### 1. Scrape the details of most viewed videos on YouTube from Wikipedia: 
    Url= https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos You need to find following details:
    A) Rank
    B) Name
    C) Artist
    D) Upload date
    E) Views

```python
url = 'https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos' #getting the required url
driver.get(url)

time.sleep(2) #wait for 2 m.s.

Rank = []
Name = []
Artist = []
Upload_date = []
Views = []

for i in range(1,31): #We have 30 rows, running the loop from 1 to 31
    Rank_tags=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr["+str(i)+"]/td[1]") #Rank
    Rank.append(Rank_tags.text) #Appending Rank
    Name_tags=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr["+str(i)+"]/td[2]") #Name
    Name.append(Name_tags.text[1:-5]) #Taking only Song Name
    Artist_tags=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr["+str(i)+"]/td[3]") #Artist
    Artist.append(Artist_tags.text) #Artist Name
    Views_tags=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr["+str(i)+"]/td[4]") #Upload_date
    Views.append(Views_tags.text) #Number of views in billions
    Upload_date_tags=driver.find_element_by_xpath("//table[@class='wikitable sortable jquery-tablesorter'][1]/tbody/tr["+str(i)+"]/td[5]") #Views
    Upload_date.append(Upload_date_tags.text) #Upload Date
    
Most_viewed_video=pd.DataFrame({}) #DataFrame
Most_viewed_video["RANK"]= Rank
Most_viewed_video["NAME"]= Name
Most_viewed_video["ARTIST"]= Artist
Most_viewed_video["UPLOAD DATE"]= Upload_date
Most_viewed_video["Views (billions)"]= Views
Most_viewed_video
```

### 2. Scrape the details team India’s international fixtures from bcci.tv. 
    Url = https://www.bcci.tv/. 
    You need to find following details: 
    A) Match title (I.e. 1st ODI) 
    B) Series 
    C) Place 
    D) Date 
    E) Time 
Note: - From bcci.tv home page you have reach to the international fixture page through code.

```python
url = 'https://www.bcci.tv/' #getting the required url
driver.get(url)

time.sleep(2) #wait for 2 m.s.

fixtures=driver.find_element_by_xpath("//div[@class='navigation__drop-down drop-down drop-down--reveal-on-hover'][1]/div/ul/li[1]/a") #Locating to International Fixtures
driver.get(fixtures.get_attribute('href')) #Opening International Fixtuers page
time.sleep(2) #wait for 2 m.s.

Match_title = []
Series = []
Place = []
Date = []
Time = []

title=driver.find_elements_by_xpath("//span[@class='u-unskewed-text fixture__tournament-label u-truncated']") #Title of the Seriers
for i in title:
    Series.append(i.text)

    
name=driver.find_elements_by_xpath("//strong[@class='fixture__name fixture__name--with-margin']") #Match Title
for i in name:
    Match_title.append(i.text)

    
loc=driver.find_elements_by_xpath("//p[@class='fixture__additional-info']/span") #Location
for i in loc:
    Place.append(i.text)

    
dt1=driver.find_elements_by_xpath("//div[@class='fixture__full-date']/span") #Day
dt2=driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[1]") #Month
dat = len(dt1)
for i in range(0,dat):
    Date.append(dt1[i].text+" "+dt2[i].text) #adding Day and Month to form Date

    
tim=driver.find_elements_by_xpath("//div[@class='fixture__date-details']/span[2]") #Time
for i in tim:
    Time.append(i.text)
    
Fixtures=pd.DataFrame({}) #DataFrame
Fixtures['Match title'] = Match_title
Fixtures['Series'] = Series
Fixtures['Place'] = Place
Fixtures['Date'] = Date
Fixtures['Time'] = Time
Fixtures
```

### 3. Scrape the details of selenium exception from guru99.com. 
    Url = https://www.guru99.com/ 
    You need to find following details: 
    A) Name 
    B) Description 
Note: - From guru99 home page you have to reach to selenium exception handling page through code.

```python
url = 'https://www.guru99.com/' #getting the required url
driver.get(url)

time.sleep(2) #wait for 2 m.s.

text1 = driver.find_elements_by_xpath("//li[@class='fa fa-chevron-circle-right']") #Looking for Selenium
text2 = driver.find_elements_by_xpath("//li[@class='fa fa-chevron-circle-right']/a") #Looking for Hyperlink
leng = len(text1)
for i in range(0,leng):
    if(text1[i].text == "Selenium"): #If selenium found open the link and break the loop
        url = text2[i].get_attribute('href')
        break
driver.get(url)
time.sleep(4)

link1 = driver.find_elements_by_xpath("//table[@class='table'][5]/tbody/tr/td[2]") #Looking for Selenium Exception Handling
link2 = driver.find_elements_by_xpath("//table[@class='table'][5]/tbody/tr/td[1]/a") #Looking for Hyperlink
leng = len(link1)
for i in range(0,leng): #If found Exception Handling than open the link and break the loop
    if(link1[i].text == "Selenium Exception Handling (Common Exceptions List)"):
        url = link2[i].get_attribute('href')
        break
driver.get(url)
time.sleep(4)

Name = []
Desc = []
for i in range(2,43):
    link1 = driver.find_element_by_xpath("//table[@class='table table-striped']/tbody/tr["+str(i)+"]/td[1]") #Exception Name
    link2 = driver.find_element_by_xpath("//table[@class='table table-striped']/tbody/tr["+str(i)+"]/td[2]") #Description
    Name.append(link1.text)
    Desc.append(link2.text)
    
Exceptions=pd.DataFrame({}) #DataFrame
Exceptions['Exception name'] = Name
Exceptions['Description'] = Desc
Exceptions
```

### 5. Scrape the details of trending repositories on Github.com. 
    Url = https://github.com/ 
    You have to find the following details: 
    A) Repository title 
    B) Repository description 
    C) Contributors count 
    D) Language used 
Note: - From the home page you have to click on the trending option from Explore menu through code.

```python
url = 'https://github.com/' #getting the required url
driver.get(url)

time.sleep(2) #wait for 2 m.s.

Trending=driver.find_element_by_xpath("//a[@href='/trending']") #Looking for Trending
driver.get(Trending.get_attribute('href')) #Open Trending page
time.sleep(2) #wait for 2 m.s.

Repository_title = []
Repository_description = []
Contributors_count = []
Language_used = []
for i in range(1,26): #25 repository are trending so running the loop for 1 to 26
    link1 = driver.find_element_by_xpath("//article[@class='Box-row']["+str(i)+"]/h1/a") #Repository Title
    Repository_title.append(link1.text)
    try:
        link2 = driver.find_element_by_xpath("//article[@class='Box-row']["+str(i)+"]/p") #Repository Description
        Repository_description.append(link2.text)
    except NoSuchElementException as e:
        Repository_description.append("-") #If Description is not present add "-"
    
urls = []
contr=driver.find_elements_by_xpath("//article[@class='Box-row']/h1/a") #Collecting the URLS of each repository
for i in contr:
    urls.append(i.get_attribute('href')) #Added url in urls field
    
for i in urls:
    driver.get(i) #Open the URL one by one
    time.sleep(2)
    counter = 0
    for i in range(1,6):
        try:
            test = driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div["+str(i)+"]/div/h2/a") #Searching for Contributors
            value = test.text
            if(value.find("Contributors") != -1): #If found than
                con = value.split(" ")            #Split value with space
                Contributors_count.append(con[1]) #Second value will be number of Contributors
            else:
                counter += 1
        except NoSuchElementException as e:
            counter += 1
            pass
    if(counter == 5):
        Contributors_count.append("-")           #If no Contributors for that Repository than add "-"
        
    count = 0
    for i in range(1,7):
        try:
            test = driver.find_element_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div["+str(i)+"]/div/h2") #Searching for Language Used
            value = test.text
            string = ""
            if(value.find("Languages") != -1): #If Language found than
                lang = driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div["+str(i)+"]/div/ul/li/a/span[1]")
                for i in lang:
                    string = string + i.text + " " #Added all languages one by one
                Language_used.append(string)
            else:
                count += 1
        except NoSuchElementException as e:
            count += 1
            pass
    if(count == 6):
        Language_used.append("-")                 #If no Language found then add "-"
        
Github = pd.DataFrame({}) #DataFrame
Github['Repository Title'] = Repository_title
Github['Repository Description'] = Repository_description
Github['Contributors Count'] = Contributors_count
Github['Language Used'] = Language_used
Github
```

### 10. Details of Datasets from UCI machine learning repositories. Url = https://archive.ics.uci.edu/ You have to find the following details:

    C) Task
    D) Attribute type
